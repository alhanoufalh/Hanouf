## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Task2t.csv')
df.head()
```

### Identify and analyze patterns in the data.

```python
Jamscountt =df['JamsCount'].value_count()
Jamscountt
```

```python
C_City = df['City'].value_counts()
C_City
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(10,6))
x= df.JamsCount
y= df.JamsDelay
plt.bar(x,y)
plt.xlabel('JamsCount')
plt.ylabel('trafficindexlive')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.figure(figsize =(10,6))
sns.scatterplot( data = df , x ='TrafficIndexLive',  y='TravelTimeHistoric')
plt.title('relation between TrafficIndexL and TravelTimeHistoric')
plt.xlabel('TrafficIndexLive')
plt.ylabel('TravelTimeHistoric')
plt.show()

```
